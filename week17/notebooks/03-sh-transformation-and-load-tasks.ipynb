```python
identify_missing_data(users_data_cleaned)
```

```python
# option 2.....

# step 1: identifying columns that contain more than 70% missing data

missing_df = identify_missing_data(users_data)
missing_df
```

```python
cols_to_drop = list(missing_df[missing_df.percent_missing > 70].variable)
cols_to_drop
```

```python
# this gives the number of column names in my list

len(cols_to_drop)
```

```python
# step 2: drop the columns

users_data_cleaned = users_data.drop(cols_to_drop, axis=1)
```

```python
# check the column has been dropped

identify_missing_data(users_data_cleaned)
```
